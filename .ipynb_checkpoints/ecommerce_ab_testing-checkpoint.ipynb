```python
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import scipy.stats as stats

events = pd.read_csv('events.csv')
events.head()
```

### Handling NaN's

```python
events.info()
```

There are 100+ NaN user_sessions, let's filer them out:

```python
events = events[events['user_session'].notna()]
```

Column "event_time" is an object, so let's make it datetime.
Also I'll add two columns for further analysis: event_month and event_dayname:

```python
# Transforming event_time to datetime
events['event_time'] = pd.to_datetime(events['event_time'], format='%Y-%m-%d %H:%M:%S UTC')

# Adding a column with a year and month for further analysis
events['event_month'] = events['event_time'].dt.strftime('%Y-%m')

# Adding a column with a week number starting from min event_time
min_date = events['event_time'].min()
first_weekday = (min_date + pd.DateOffset(days=(7 - min_date.weekday()))).normalize()
events['event_weeknum'] = ((events['event_time'] - first_weekday).dt.days // 7) + 1

# Adding a column with weekday names for further analysis
events['event_dayname'] = events['event_time'].dt.day_name()
events.head()
```

### Conversion analysis

Now let's see if the proportion of views/additions to cart/purchases changes over 5 months 
(not including the last week of September):


```python
# Calculating shares of event types in every month
events_test = events[events['event_month'] != '2020-09'].groupby('event_month', as_index=False)['event_type'].value_counts(normalize=True)
events_test['proportion'] = round(events_test['proportion'], 2)

# Plotting
ax = sns.barplot(data=events_test, x='event_month', y='proportion', hue='event_type', color='blue')

# Show labels
for i in ax.containers:
    ax.bar_label(i,)
```

It looks like **there is a slow trend of increasing additions to cart and purchases and decreasing views over these months**.
We can see that more clearly by calculating conversions:

```python
event_types_per_month = events[events['event_month'] != '2020-09'].groupby('event_month', as_index=False)['event_type'].value_counts()
conversions = pd.DataFrame(columns=['event_month','view/cart','cart/purchase','view/purchase'])

for month in event_types_per_month['event_month'].unique().tolist():
    cr_view_cart = event_types_per_month[(event_types_per_month['event_month'] == month) & (event_types_per_month['event_type'] == 'cart')]['count'].sum() / event_types_per_month[(event_types_per_month['event_month'] == month) & (event_types_per_month['event_type'] == 'view')]['count'].sum()
    cr_cart_purchase = event_types_per_month[(event_types_per_month['event_month'] == month) & (event_types_per_month['event_type'] == 'purchase')]['count'].sum() / event_types_per_month[(event_types_per_month['event_month'] == month) & (event_types_per_month['event_type'] == 'cart')]['count'].sum()
    cr_view_purchase = event_types_per_month[(event_types_per_month['event_month'] == month) & (event_types_per_month['event_type'] == 'purchase')]['count'].sum() / event_types_per_month[(event_types_per_month['event_month'] == month) & (event_types_per_month['event_type'] == 'view')]['count'].sum()
    conversions = conversions.append({'event_month': month, 'view/cart': cr_view_cart, 'cart/purchase': cr_cart_purchase, 'view/purchase': cr_view_purchase}, ignore_index=True)

conversions
    
```

Now I'll plot the conversions to see the differences more clear:

```python
plt.figure(figsize=(8,3))

plt.plot(conversions['event_month'], conversions['view/cart'], label='View/Cart', marker='o')
plt.plot(conversions['event_month'], conversions['cart/purchase'], label='Cart/Purchase', marker='o')
plt.plot(conversions['event_month'], conversions['view/purchase'], label='View/Purchase', marker='o')

plt.ylabel('Conversion Rate')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show();
```

1. We can clearly see that **conversions view/cart and view/purchase have been increasing for 5 months straight**
2. Conversion **cart/purchase**, on the other hand, **has been declining for 4 months but showed a slight increase on the 5th month**
3. Conversion **cart/purchase is very high, meaning people who add a product to cart then purchase this product with 60+% probability!**

It's now interesting to see if there are any differences in conversions between the weekdays:

```python
event_types_per_day = events.groupby('event_dayname', as_index=False)['event_type'].value_counts()
conversions_daily = pd.DataFrame(columns=['event_dayname','view/cart','cart/purchase','view/purchase'])

for day in event_types_per_day['event_dayname'].unique().tolist():
    cr_view_cart = event_types_per_day[(event_types_per_day['event_dayname'] == day) & (event_types_per_day['event_type'] == 'cart')]['count'].sum() / event_types_per_day[(event_types_per_day['event_dayname'] == day) & (event_types_per_day['event_type'] == 'view')]['count'].sum()
    cr_cart_purchase = event_types_per_day[(event_types_per_day['event_dayname'] == day) & (event_types_per_day['event_type'] == 'purchase')]['count'].sum() / event_types_per_day[(event_types_per_day['event_dayname'] == day) & (event_types_per_day['event_type'] == 'cart')]['count'].sum()
    cr_view_purchase = event_types_per_day[(event_types_per_day['event_dayname'] == day) & (event_types_per_day['event_type'] == 'purchase')]['count'].sum() / event_types_per_day[(event_types_per_day['event_dayname'] == day) & (event_types_per_day['event_type'] == 'view')]['count'].sum()
    conversions_daily = conversions_daily.append({'event_dayname': day, 'view/cart': cr_view_cart, 'cart/purchase': cr_cart_purchase, 'view/purchase': cr_view_purchase}, ignore_index=True)


weekdays_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
conversions_daily = conversions_daily.groupby(['event_dayname']).sum().reindex(weekdays_order).reset_index()

# conversions_daily
plt.figure(figsize=(8,3))

plt.plot(conversions_daily['event_dayname'], conversions_daily['view/cart'], label='View/Cart', marker='o')
plt.plot(conversions_daily['event_dayname'], conversions_daily['cart/purchase'], label='Cart/Purchase', marker='o')
plt.plot(conversions_daily['event_dayname'], conversions_daily['view/purchase'], label='View/Purchase', marker='o')

plt.ylabel('Conversion Rate')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(0, 0.75)

plt.show();
    
```

### Retention analysis

Considering that **we only have data for 5 months** and **we are dealing with users' behavior in a large ecommerce store
(which means that users don't visit the store frequently)**, I want to take a look at **weekly retention**.


```python
# Finding the first activity week for each user_id
first_activity_week = events[events['event_weeknum'] != 0].groupby('user_id')['event_weeknum'].min().reset_index()
first_activity_week.rename(columns={'event_weeknum': 'cohort_week'}, inplace=True)

# Merging first week with the main dataframe
retention_data = pd.merge(events[events['event_weeknum'] != 0][['event_time', 'user_id', 'event_weeknum']], first_activity_week, on='user_id')

# Calculating the difference (in weeks) between the first activity week and a certain week
retention_data['weeks_since_first_activity'] = retention_data['event_weeknum'] - retention_data['cohort_week']

# Calculating retention
def calculate_retention(retention_data):
    cohort_data = retention_data.groupby(['cohort_week', 'weeks_since_first_activity']).agg(n_users=('user_id', 'nunique')).reset_index()

    cohort_sizes = cohort_data[cohort_data['weeks_since_first_activity'] == 0].rename(columns={'n_users': 'cohort_size'})
    cohort_data = cohort_data.merge(cohort_sizes[['cohort_week', 'cohort_size']], on='cohort_week')

    cohort_data['retention'] = cohort_data['n_users'] / cohort_data['cohort_size']

    retention = cohort_data.pivot(index='cohort_week', columns='weeks_since_first_activity', values='retention') \
        .reset_index() \
        .rename_axis(None, axis=1)

    return retention

retention = calculate_retention(retention_data)
retention.head(10)
```

```python
import plotly.express as px

retention_melted = retention.melt(id_vars='cohort_week', var_name='weeks_since_first_activity', value_name='retention')
retention_melted['retention'] = retention_melted['retention'] * 100

# Plotting
fig = px.line(retention_melted, x='weeks_since_first_activity', y='retention', color='cohort_week',
              title='Retention by Cohort Over Time',
              labels={'weeks_since_first_activity': 'Weeks Since First Activity', 'retention': 'Retention (%)'},
              range_y=[0, 4])

fig.show();
```

```python
# Now let's calculate average n-week retention for a clearer image

# Calculating average retention rate for each number of weeks
avg_retention_by_week = retention.drop(columns=['cohort_week']).apply(lambda x: x.dropna().mean())

avg_retention = avg_retention_by_week.reset_index()
avg_retention.columns = ['weeks_since_first_activity', 'avg_retention']
avg_retention['avg_retention'] = avg_retention['avg_retention'] * 100

# Plotting
plt.figure(figsize=(8, 3))
sns.lineplot(data=avg_retention, x='weeks_since_first_activity', y='avg_retention', marker='o')

plt.xlabel('Weeks Since First Activity')
plt.ylabel('Average Retention (%)')
plt.ylim(0, 4)
# plt.yticks([range(0, 4, 0.5)])
plt.xticks(avg_retention['weeks_since_first_activity'])

plt.show();
```

Now I want to analyse the rolling retention for better understanding.

**For example, 1-week rolling retention is:**
- We take users who were active in the store last week
- We calculate what share of them is also active this week


```python
def calculate_rolling_retention(data, intervals):

    weeks = data['event_weeknum'].unique()
    
    rolling_retention = pd.DataFrame()

    for week in weeks:
        total_users = data[data['event_weeknum'] == week]['user_id'].unique()
        total_users_cnt = len(total_users)

        week_data = {'cohort_week': week, 'total_users': total_users_cnt}

        for interval in intervals:
            users_per_interval = data[(data['event_weeknum'] > week) & 
                                      (data['event_weeknum'] <= week + interval) & 
                                      (data['user_id'].isin(total_users))]['user_id'].nunique()
            
            week_data[f'retained_in_{interval}_weeks'] = users_per_interval
            week_data[f'retention_{interval}_weeks'] = users_per_interval / total_users_cnt * 100

        rolling_retention = rolling_retention.append(week_data, ignore_index=True)
            
    return rolling_retention

weeks_intervals = [1, 4, 8, 12]

rolling_retention = calculate_rolling_retention(events[events['event_weeknum'] != 0], weeks_intervals)
rolling_retention = rolling_retention[['cohort_week','total_users',
                                       'retained_in_1_weeks','retained_in_4_weeks','retained_in_8_weeks','retained_in_12_weeks',
                                       'retention_1_weeks','retention_4_weeks','retention_8_weeks','retention_12_weeks']]
rolling_retention.head()
```

```python
# Plotting
plt.figure(figsize=(8, 3))

for interval in weeks_intervals:
    max_week = rolling_retention['cohort_week'].max() - interval
    sns.lineplot(data=rolling_retention[rolling_retention['cohort_week'] <= max_week], x='cohort_week', y=f'retention_{interval}_weeks', marker='o', label=f'{interval} weeks')

plt.xlabel('Cohort (Week Number)')
plt.ylabel('Rolling Retention (%)')
plt.ylim(3, 8)
plt.xticks(rolling_retention['cohort_week'])

plt.show()
```

**Here we can see:**
- **1-week rolling retention is around 4% on average with a slight increase in the last weeks**
- The **higher the rolling window** - the **higher is rolling retention**, 
    because users are not visiting the store frequently

## Experiment design

Now let's use the data we have as historical data to design an A/B experiment.

### The idea of the experiment

After looking at the conversion rates the company decided to find and test some hypothesis to increase the conversion
from view to addition to cart. 

### Hypothesis

- **changing the design** of the product page will **increase the probability of adding a product to cart**, 
    which will be evaluated by the **view/cart conversion**.

Now let's design the experiment and start with alpha, beta, and MDE.

### Probabilities of type I and type II errors, MDE

Firstly, I'm going to create a dataframe where for **each** user 
I can aggregare their views/carts/purchases and show their segments (this dataframe will be used for post-stratification):


**We are assuming that today is February 21, 2021, so I'll use the week from February 15 to February 21 for experiment design,
and the last week in the dataset (from February 22 to February 28) - as the experiment week.**

Designing includes different steps, including the selection of probabilities of type I and type II errors, MDE, sample size, etc.

I'll use the formula of sample size calculation to calculate the MDE we can detect with the traffic we get a week:

$$  n > \frac{[\Phi^{-1}(1- \alpha)/2) + \Phi^{-1}(1 - \beta)]^2(\sigma_X^2 + \sigma_Y^2)}{\epsilon^2} $$

$$  \epsilon > \sqrt{\frac{[\Phi^{-1}(1- \alpha)/2) + \Phi^{-1}(1 - \beta)]^2(\sigma_X^2 + \sigma_Y^2)}{n}} $$

But before calculating the sample size, we need to do the **linearization** (transformation from a ratio metric
to a user metric)


### Linearization

In this experiment we are working with a **conversion**, which is a **ratio metric**.

**Ratio metrics** are a lot harder to work on, because we only have **one metric value for the whole group of users**.
It's also a lot more difficult to properly design an experiment, calculate the sample size and MDE.

That is why **linearization is used, which is just a way to transform a ratio metric to a user metric (1 value for each user)**.
When we have a linearized user metric, we can easily design an experiment and implement post-stratification or
other variance reduction methods.

**The formula for a linearized metric looks like this:**

$$ L(u) = X(u) - R_A Y(u), \forall{u} \in A $$

$u$ - user, 

$R_A$ - ratio metric of group A, 

$X(u)$ - numerator (in our case, total number of "cart"s among all users),

$Y(u)$ - denominator (in our case, total number of "view"s among all users)

**Before we start, let's check how many unique users we get each week (excluding the experiment week):**


```python
events[events['event_weeknum'].isin(range(1, 22))].groupby('event_weeknum', as_index=False)['user_id'].nunique()
```

Okay, **let's pick 17500 unique users, because if we design the experiment using 19000 users and next week we get only 17000 
(as an example), the experiment design will be incorrect**.

Now let's start by creating a separate dataframe with users who interacted with the website 
**during the week before the last one (because I'll use the last week as an experiment week)**:

```python
# Getting a separate dataframe for only last week's 17500 users
last_week_data_per_user = events[events['event_weeknum'] == 21] \
    .pivot_table(index='user_id', columns='event_type', values='event_time', aggfunc='count', fill_value=0) \
    .reset_index() \
    .rename_axis(None, axis=1)

last_week_data_per_user = pd.merge(
    last_week_data_per_user,
    events[events['event_weeknum'] == 21][['user_id','segment_purchase','segment_category','segment_price']].drop_duplicates(),
    how='left',
    on='user_id'
)

# Getting 17500 random users from all last week's users
last_week_users = random.sample(last_week_data_per_user['user_id'].unique().tolist(), 17500)

# Filtering data
last_week_data_per_user = last_week_data_per_user[last_week_data_per_user['user_id'].isin(last_week_users)]
```

Now I'll calculate the **ratio metric** and apply the **linearization**:

```python
# Calculating conversion 
cr_view_cart = last_week_data_per_user['cart'].sum() / last_week_data_per_user['view'].sum()

# Adding a column with a linearized metric for each user
last_week_data_per_user['lin_view_cart'] = last_week_data_per_user['cart'] - cr_view_cart * last_week_data_per_user['view']
last_week_data_per_user
```

```python
# Calculating mean and std
mean = round(np.mean(last_week_data_per_user['lin_view_cart']), 4)
std = round(np.std(last_week_data_per_user['lin_view_cart']), 4)

print(f'mean = {mean}, std = {std}')
```

```python
alpha = 0.05
beta = 0.2

def get_minimal_determinable_effect(std, sample_size, alpha, beta):
    t_alpha = stats.norm.ppf(1 - alpha / 2, loc=0, scale=1)
    t_beta = stats.norm.ppf(1 - beta, loc=0, scale=1)
    disp_sum_sqrt = (2 * (std ** 2)) ** 0.5
    mde = (t_alpha + t_beta) * disp_sum_sqrt / np.sqrt(sample_size)
    return mde

mde = get_minimal_determinable_effect(std, len(last_week_data_per_user) // 2, alpha, beta)

print(f'''
mde linearized: {round(mde, 4)}
mde linearized (in %): {np.round(mde / mean * 100, 3)}%
sample size: {len(last_week_data_per_user) // 2}
''')
```

So, if we use **all** traffic we get for 2 weeks, we can detect an effect of 0.024,
which is 142.246% change of the mean!

This is very high, so let's see if we can increase the sensitivity of the test by removing outliers.

To do that, let's plot the distribution of the linearized metric:


### CUPED

The MDE is still too high, so let's try to reduce variance using CUPED method.

To do that, I'll take the users from the previous week and calculate the linearized metric for each of them.

```python
# Getting a separate dataframe for pre-last week data
pre_last_week_data_per_user = events[events['event_weeknum'] == 20] \
    .pivot_table(index='user_id', columns='event_type', values='event_time', aggfunc='count', fill_value=0) \
    .reset_index() \
    .rename_axis(None, axis=1)

pre_last_week_data_per_user = pd.merge(
    pre_last_week_data_per_user,
    events[events['event_weeknum'] == 20][['user_id','segment_purchase','segment_category','segment_price']].drop_duplicates(),
    how='left',
    on='user_id'
)

# Calculating conversion in pre-design data
cr_view_cart_pre = pre_last_week_data_per_user['cart'].sum() / pre_last_week_data_per_user['view'].sum()

# Calculating linearized metric for each user in pre-design period
pre_last_week_data_per_user['lin_view_cart'] = pre_last_week_data_per_user['cart'] - cr_view_cart_pre * pre_last_week_data_per_user['view']
pre_last_week_data_per_user
```

Now we need to join the data from both periods - design and pre-design:
    

```python
cuped_design = pd.merge(
    # last_week_data_per_user[['user_id','lin_view_cart','segment_purchase','segment_category','segment_price']].rename(columns={'lin_view_cart': 'lin_metric'}),
    last_week_data_per_user.rename(columns={'lin_view_cart': 'lin_metric'}),
    pre_last_week_data_per_user[['user_id','lin_view_cart']].rename(columns={'lin_view_cart': 'lin_metric_prev'}),
    how='left',
    on='user_id'
)

cuped_design['lin_metric_prev'] = cuped_design['lin_metric_prev'].fillna(0)

# Calculating correlation between the linearized metric and covariate 
# (which in this case is linearized metric during previous week)
cuped_design[['lin_metric','lin_metric_prev']].corr()
```

The **conversion between the linearized metric and covariate is pretty low, 
which means that the variance can not be reduced significantly**, unfortunately, but I'll finish the CUPED.

Now we need to transform the linearized metric to CUPED metric, 
and to do that we first need to calculate $\theta$:


```python
def calculate_theta(metric, covariate):
    covariance = np.cov(covariate, metric)[0, 1]
    variance = covariate.var()
    theta = covariance / variance
    return theta

a_users, b_users = np.random.choice(cuped_design['user_id'], (2, len(cuped_design) // 2,), False)

# Calculating theta
theta = calculate_theta(cuped_design['lin_metric'], cuped_design['lin_metric_prev'])

# Calculating CUPED metric
cuped_design['cuped_metric'] = cuped_design['lin_metric'] - theta * cuped_design['lin_metric_prev']

# Calculating variance of the CUPED metric
cuped_var = np.var(cuped_design['cuped_metric'])
cuped_mean = np.mean(cuped_design['cuped_metric'])

print(f'''
linearized metric variance = {std**2}, mean = {mean} 
cuped metric variance = {cuped_var}, mean = {cuped_mean}
decrease = x{round(std**2 / cuped_var,1)}
''')
```

```python
alpha = 0.05
beta = 0.2

mde_cuped = get_minimal_determinable_effect(np.sqrt(cuped_var), len(cuped_design) // 2, alpha, beta)

print(f'''
mde cuped: {round(mde_cuped, 4)}
mde cuped (in %): {np.round(mde_cuped / cuped_mean * 100, 3)}%
---------------------------
mde linearized: {round(mde, 4)}
mde linearized (in %): {np.round(mde / mean * 100, 3)}%
---------------------------
mde decrease = {round(mde / mde_cuped, 1)}
---------------------------
sample size: {len(cuped_design) // 2}
''')
```

### Removing outliers

Now let's see if we can reduce the variance even more by removing outliers.

We need to plot the CUPED metric distribution first:


```python
# Plotting
fig, axs = plt.subplots(1,2,figsize=(10,4))

ax1 = sns.histplot(cuped_design['cuped_metric'], bins=100, ax=axs[0])
ax2 = sns.histplot(last_week_data_per_user['lin_view_cart'], bins=100, ax=axs[1])

ax1.set_ylim(0, 100)
ax2.set_ylim(0, 100)
ax1.set_title('Distribution of CUPED metric')
ax2.set_title('Distribution of linearized metric')

# Quantiles
quantiles_values = [0.25, 99.75]

quantiles_cuped = np.percentile(cuped_design['cuped_metric'], quantiles_values)
quantiles_lin = np.percentile(last_week_data_per_user['lin_view_cart'], quantiles_values)

for q in quantiles_cuped:
    ax1.axvline(q, color='red', linestyle='--')

for q in quantiles_lin:
    ax2.axvline(q, color='red', linestyle='--')
    
# Adding legend with quantiles
ax1.legend(labels=[f'{quantiles_values[0]}% quantile = {quantiles_cuped[0]:.2f}', f'{quantiles_values[1]}% quantile = {quantiles_cuped[1]:.2f}'])
ax2.legend(labels=[f'{quantiles_values[0]}% quantile = {quantiles_lin[0]:.2f}', f'{quantiles_values[1]}% quantile = {quantiles_lin[1]:.2f}']);
```

```python
# Creating a dataset without outliers (we're removing 0.5% of data from each tail of the distribution)
cuped_design_no_outliers = cuped_design[(cuped_design['cuped_metric'] >= -1.31) & (cuped_design['cuped_metric'] <= 3.51)]

# Mean and std
mean_cuped_no_outliers = round(np.mean(cuped_design_no_outliers['cuped_metric']), 4)
std_cuped_no_outliers = round(np.std(cuped_design_no_outliers['cuped_metric']), 4)

print(f'''
linearized metric mean = {mean}, linearized metric variance = {std**2},  
CUPED metric mean = {cuped_mean}, CUPED metric variance = {cuped_var},
CUPED mean (no outliers) = {mean_cuped_no_outliers}, CUPED variance (no outliers) = {std_cuped_no_outliers**2},
------------------------------------------------------------------------------
variance decrease = x{round((std**2) / (std_cuped_no_outliers**2), 1)}
------------------------
sample_size = {len(cuped_design_no_outliers) // 2}
''')
```

The variance after removing outliers decreased even more!

Let's see how that affects the MDE for the same sample size:

```python
alpha = 0.05
beta = 0.2

mde_cuped_no_outliers = get_minimal_determinable_effect(std_cuped_no_outliers, len(cuped_design_no_outliers) // 2, alpha, beta)

print(f'''
mde cuped (no outliers): {round(mde_cuped_no_outliers, 4)}
mde cuped (no outliers) (in %): {np.round(mde_cuped_no_outliers / mean_cuped_no_outliers * 100, 3)}%
------------------------------------------------------
mde cuped: {round(mde_cuped, 4)}
mde cuped (in %): {np.round(mde_cuped / cuped_mean * 100, 3)}%
------------------------------------------------------
mde linearized: {round(mde, 4)}
mde linearized (in %): {np.round(mde / mean * 100, 3)}%
------------------------------------------------------
mde decrease (linearized -> cuped w/o outliers) = x{round(mde / mde_cuped_no_outliers, 1)}
------------------------------------------------------
sample size: {len(cuped_design_no_outliers) // 2}
''')
```

Okay, now we have **MDE in the CUPED metric's scale**,
so **we need to convert this effect into the ratio metric to make it easier to understand by the business**.

I will use this formula:
$ dR = \frac{dL}{\frac{1}{|B|} \sum_{u \in B} Y(u)} $

$dR$ - absolute effect in ratio metric,
$dL$ - absolute effect in linearized metric,
$\frac{1}{|B|} \sum_{u \in B} Y(u)$ - mean of views


```python
# Initial absolute MDE in the ratio metric scale
mde_ratio_initial = mde / np.mean(last_week_data_per_user['view'])

# Initial relative MDE in the ratio metric scale
mde_ratio_relative_initial = mde_ratio_initial / cr_view_cart

# Absolute MDE in the ratio metric scale
mde_ratio = mde_cuped_no_outliers / np.mean(cuped_design_no_outliers['view'])
cr_view_cart_cuped_no_outliers = cuped_design_no_outliers['cart'].sum() / cuped_design_no_outliers['view'].sum()

# Relative MDE in the ratio metric scale
mde_ratio_relative = mde_ratio / cr_view_cart_cuped_no_outliers

print(f'''
Initial:
- mde of initial ratio metric: {round(mde_ratio_initial, 4)}
- mde of initial ratio metric (in %): {np.round(mde_ratio_relative_initial * 100, 3)}%
- sample size: {len(last_week_data_per_user) // 2}
----------------------------------------------------------
After linearizarion + CUPED + removing outliers:
- mde of ratio metric after CUPED and removing outliers: {round(mde_ratio, 4)}
- mde of ratio metric after CUPED and removing outliers (in %): {np.round(mde_ratio_relative * 100, 3)}%
- sample size: {len(cuped_design_no_outliers) // 2}
----------------------------------------------------------
Ratio metric MDE decrease = x{round(mde_ratio_initial / mde_ratio, 1)}
''')
```

Good, it looks like if we: 
- **divide 17500 users we get next week** into two groups (which 8750 users per group),
- apply **linearization** to get a user metric instead of working with ratio metric,
- then apply **CUPED** method to get a new CUPED metric based on the lineraized metric,
- **remove 0.5%** from each tail of the CUPED metric's distribution,

**we can detect a 1.4x smaller effect of the conversion with the same $\alpha = 0.05$ and $\beta = 0.2$**!

## Design testing - synthetic A/A and A/B experiments

To make sure that:
- the probability of type I error is controlled at level $\alpha = 0.05$
- the probability of type II error is controlled at level $\beta = 0.2$

We need to run 1000 (or more) synthetic A/A and A/B experiments, 
in each of them we randomly select group A and B with a size = calculated sample size.

- In A/A experiments there is no effect in the second group: null hypothesis ($H_0$) is true
- In A/B experiments there is an effect = calculated effect in the second group: null hypothesis ($H_0$) is false

For each synthetic A/A and A/B experiment we calculate the p-value and then plot it:
- For A/A tests p-value should have a uniform distribution
- For A/B tests p-value should have a convex distribution


```python
def plot_pvalue_ecdf(pvalues_aa, pvalues_ab):
    
    fig, ax = plt.subplots(1, 2, figsize=(10, 3))

    ax1 = sns.ecdfplot(pvalues_aa, ax=ax[0])
    ax1.plot([0,1],[0,1], 'k--')
    ax1.set(xlabel='p-value', ylabel='Probability', title='p-value ECDF: A/A')
    ax1.grid(True)
    
    ax2 = sns.ecdfplot(pvalues_ab, ax=ax[1])
    ax2.plot([0,1],[0,1], 'k--')
    ax2.set(xlabel='p-value', ylabel='Probability', title='p-value ECDF: A/B')
    ax2.grid(True)

p_values_aa = []
p_values_ab = []
for _ in range(1000):
    a1, a2 = np.random.choice(cuped_design_no_outliers['cuped_metric'], (2, len(cuped_design_no_outliers) // 2), False)
    p_value_aa = stats.ttest_ind(a1, a2).pvalue
    p_values_aa.append(p_value_aa)
    
    a, b = np.random.choice(cuped_design_no_outliers['cuped_metric'], (2, len(cuped_design_no_outliers) // 2), False)
    b += mde_cuped_no_outliers # adding the effect
    p_value_ab = stats.ttest_ind(a, b).pvalue
    p_values_ab.append(p_value_ab)

plot_pvalue_ecdf(p_values_aa, p_values_ab)
```

```python
# Caluculating the power of the test

p_values_binary = [1 if p < 0.05 else 0 for p in p_values_ab]
np.mean(p_values_binary)
```

- p-value distribution of 1000 synthetic A/A experimtns is **uniform**
- p-value distribution of 1000 synthetic A/B experimtns is **convex with probability = 0.8 for p-value = 0.05**
    (which means that if there is an effect in group B, the test detects it (p-value >= 0.05) with 80% probability)

**Looks perfect!**

### Post-stratification

**Post-stratification** is a technique that is used for **variance reduction**. 

The idea of post-stratification is to **reweight the stratas/segments in groups A and B 
so that their weights are equal to the weights of the stratas/segments in the general population**

My first idea was to implement **stratification, which is a more complex technique that requires to do the stratified sampling**,
however, with the given dataset **it is not possible due to low retention rate and the fact that most users who will
participate in the experiment will be new users whose segment we don't know yet**.

**Post-stratification**, on the other hand, only requires to do the simple random sampling (SRS) 
and then calculate stratified means and variances in each group, 
but works almost as good as normal stratification in terms of variance reduction.

Let's start by calculating **weights of segments in population**:


```python
# Getting a dataframe with all users and their views/carts/purchases and segments over all 5 months 
# (excluding the first few days)

all_data_per_user = events[events['event_weeknum'] != 0] \
    .pivot_table(index='user_id', columns='event_type', values='event_time', aggfunc='count', fill_value=0) \
    .reset_index() \
    .rename_axis(None, axis=1)

all_data_per_user = pd.merge(
    all_data_per_user,
    events[events['event_weeknum'] != 0][['user_id','segment_purchase','segment_category','segment_price']].drop_duplicates(),
    how='left',
    on='user_id'
)

all_data_per_user
```

```python
# Weights of segments in population
weights_segment_purchase = dict(all_data_per_user['segment_purchase'].value_counts(normalize=True))

weights_segment_category = dict(all_data_per_user['segment_category'].value_counts(normalize=True))

weights_segment_price = dict(all_data_per_user['segment_price'].value_counts(normalize=True))

print(weights_segment_purchase)
print(weights_segment_category)
print(weights_segment_price)
```

```python
# Defining necessary functions for post-stratification

def calculate_stratified_mean(df, weights):
    """  .

    df -        
    weights -  - { :  }

    return: float, mean_strat
    """
    strat_mean = df.groupby('strat')['metric'].mean()
    df_means_weights = pd.merge(
        strat_mean,
        pd.Series(weights, name='weight'),
        how='inner',
        left_index=True,
        right_index=True
    )
    #    df   - ,   weights       1
    df_means_weights['weight'] = df_means_weights['weight'] / df_means_weights['weight'].sum()
    mean_strat = (df_means_weights['weight'] * df_means_weights['metric']).sum()
    return mean_strat

def calculate_stratified_var(df, weights):
    """  ."""
    strat_vars = df.groupby('strat')['metric'].var()
    df_vars_weights = pd.merge(
        strat_vars,
        pd.Series(weights, name='weight'),
        how='inner',
        left_index=True,
        right_index=True
    )
    #    df   - ,   weights       1
    df_vars_weights['weight'] = df_vars_weights['weight'] / df_vars_weights['weight'].sum()
    var_strat = (df_vars_weights['weight'] * df_vars_weights['metric']).sum()
    return var_strat

def check_stratified_test(df_control, df_pilot, weights):
    """       .
    
    return - pvalue.
    """
    mean_strat_control = calculate_stratified_mean(df_control, weights)
    mean_strat_pilot = calculate_stratified_mean(df_pilot, weights)
    var_strat_control = calculate_stratified_var(df_control, weights)
    var_strat_pilot = calculate_stratified_var(df_pilot, weights)
    delta_mean_strat = mean_strat_pilot - mean_strat_control
    std_mean_strat = (var_strat_pilot / len(df_pilot) + var_strat_control / len(df_control)) ** 0.5
    t = delta_mean_strat / std_mean_strat
    pvalue = (1 - stats.norm.cdf(np.abs(t))) * 2
    return pvalue
```

Now let's calculate simple and stratified means and variances using different user segmentation methods
to choose the best one.

I'll do this on group A, where there is no effect:


```python
# Simple mean and var
simple_mean = round(np.mean(cuped_a_no_outliers['lin_metric']), 4)
simple_var = round(np.var(cuped_a_no_outliers['lin_metric']), 4)

# Stratified mean based on segmenting by purchase during the first session
mean_segment_purchase = round(calculate_stratified_mean(
    cuped_a_no_outliers[['segment_purchase','cuped_metric']] \
        .rename(columns={'segment_purchase': 'strat', 'cuped_metric': 'metric'}), 
    weights_segment_purchase), 4)

# Stratified mean based on segmenting by most looked category during the first session
mean_segment_category = round(calculate_stratified_mean(
    cuped_a_no_outliers[['segment_category','cuped_metric']] \
        .rename(columns={'segment_category': 'strat', 'cuped_metric': 'metric'}), 
    weights_segment_category), 4)

# Stratified mean based on segmenting by mean price of products browsed during the first session
mean_segment_price = round(calculate_stratified_mean(
    cuped_a_no_outliers[['segment_price','cuped_metric']] \
        .rename(columns={'segment_price': 'strat', 'cuped_metric': 'metric'}), 
    weights_segment_price), 4)

# Stratified variance based on segmenting by purchase during the first session
var_segment_purchase = round(calculate_stratified_var(
    cuped_a_no_outliers[['segment_purchase','cuped_metric']] \
        .rename(columns={'segment_purchase': 'strat', 'cuped_metric': 'metric'}), 
    weights_segment_purchase), 4)

# Stratified variance based on segmenting by most looked category during the first session
var_segment_category = round(calculate_stratified_var(
    cuped_a_no_outliers[['segment_category','cuped_metric']] \
        .rename(columns={'segment_category': 'strat', 'cuped_metric': 'metric'}), 
    weights_segment_category), 4)

# Stratified variance based on segmenting by mean price of products browsed during the first session
var_segment_price = round(calculate_stratified_var(
    cuped_a_no_outliers[['segment_price','cuped_metric']] \
        .rename(columns={'segment_price': 'strat', 'cuped_metric': 'metric'}), 
    weights_segment_price), 4)

print(f'''
simple mean = {simple_mean}, simple variance = {simple_var}
strat mean (purchase) = {mean_segment_purchase}, strat variance (purchase) = {var_segment_purchase},
strat mean (category) = {mean_segment_category}, strat variance (category) = {var_segment_category},
strat mean (price) = {mean_segment_price}, strat variance (price) = {var_segment_price},
''')
```

### Confidence interval with Bootstrap

Finally, let's construct a 90% and a 95% confidence intervals for the effect using Bootsrap.

To do so, I will calculate the effect between stratified means of groups A and B:


```python
def get_difference(a, b):
    mean_a = calculate_stratified_mean(
        a[['segment_purchase','cuped_metric']] \
            .rename(columns={'segment_purchase': 'strat', 'cuped_metric': 'metric'}), 
            weights_segment_purchase)
    
    mean_b = calculate_stratified_mean(
        b[['segment_purchase','cuped_metric']] \
            .rename(columns={'segment_purchase': 'strat', 'cuped_metric': 'metric'}), 
            weights_segment_purchase)
    
    return mean_b - mean_a

real_effect_cuped = get_difference(cuped_a_no_outliers, cuped_b_no_outliers)
print(f'effect = {real_effect_cuped}')

bs_point_estimates = []
for _ in range(1000):
    bs_a = cuped_a_no_outliers.sample(n=len(cuped_a_no_outliers), replace=True)
    bs_b = cuped_b_no_outliers.sample(n=len(cuped_b_no_outliers), replace=True)
    bs_point_estimate = get_real_effect(bs_a, bs_b)
    bs_point_estimates.append(bs_point_estimate)


# Plotting the distribution of differences between groups
plt.figure(figsize=(8,3))
plt.hist(bs_point_estimates, bins=20);
```

Let's check this distribution for normality to understand if I can use the normal CI:


```python
pvalue_bootstrap = stats.shapiro(bs_point_estimates).pvalue

if pvalue_bootstrap < 0.05:
    print(f'Reject H0 - the sample does not come from normal distribution (p-value = {round(pvalue_bootstrap, 4)} < 0.05)')
else:
    print(f'Can not reject H0 - the sample comes from normal distribution (p-value = {round(pvalue_bootstrap, 4)} >= 0.05)')
    
```

```python
# The distribution is normal -> we can use normal CI

def get_ci_bootstrap_normal(boot_metrics, pe_metric, alpha):
    """   .

    boot_metrics -  ,    
    pe_metric -   
    alpha -  
    
    return: (left, right) -   .
    """
    c = stats.norm.ppf(1 - alpha / 2)
    se = np.std(boot_metrics)
    left, right = pe_metric - c * se, pe_metric + c * se
    return [left, right]

# CI for CUPED metric
ci_95_cuped = get_ci_bootstrap_normal(bs_point_estimates, real_effect_cuped, 0.05)
ci_90_cuped = get_ci_bootstrap_normal(bs_point_estimates, real_effect_cuped, 0.1)

ci_95_cuped = [round(i, 4) for i in ci_95_cuped]
ci_90_cuped = [round(i, 4) for i in ci_90_cuped]

# CI for conversion
real_effect_ratio = real_effect_cuped / np.mean(cuped_a_no_outliers['view'])
cr_cuped_no_outliers = cuped_a_no_outliers['cart'].sum() / cuped_a_no_outliers['view'].sum()
real_effect_ratio_relative = effect_ratio / cr_cuped_no_outliers * 100

ci_95_cr = [round(i / np.mean(cuped_a_no_outliers['view']), 4) for i in ci_95_cuped]
ci_90_cr = [round(i / np.mean(cuped_a_no_outliers['view']), 4) for i in ci_90_cuped]

print(f'''
CUPED effect = {round(real_effect_cuped, 4)}
95% CI for effect (CUPED metric): {ci_95_cuped}
90% CI for effect (CUPED metric): {ci_90_cuped}
--------------------------------------------------
conversion effect = {round(real_effect_ratio, 4)}
conversion effect (relative) = {round(real_effect_ratio_relative, 4)}%
95% CI for effect (conversion): {ci_95_cr}
90% CI for effect (conversion): {ci_90_cr}
''')
```
